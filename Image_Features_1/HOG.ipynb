```python
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure

#    
image = cv2.imread('image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# HOG  
fd, hog_image = hog(gray, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True)

#  
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
plt.imshow(hog_image_rescaled, cmap='gray')
plt.title('HOG Features')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

#  
image = cv2.imread('ped.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# HOG   
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

#  
(rects, weights) = hog.detectMultiScale(gray, winStride=(1, 1), padding=(100, 10), scale=1.05)

#    
for (x, y, w, h) in rects:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

#  
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('HOG Detection')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

#   
video_path = 'vtest.avi'
cap = cv2.VideoCapture(video_path)

# HOG   
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    #  
    (rects, weights) = hog.detectMultiScale(gray, winStride=(8, 8), padding=(8, 8), scale=1.05)

    #    
    for (x, y, w, h) in rects:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    #  
    cv2.imshow('HOG Detection', frame)

    # 'q'   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
